```python
import imageio
import os
import numpy as np
import cv2
```

# Load stimuli data

```python
CAT_STIMULI_PATH = "/home/gabir/DATASETS/CER01A50/Stimuli central"
```

```python
def load_images_stimuli(stimuli_path):
    all_stimuli=[]
    for dir_id in range(1,4):
        stimuli = []
        stim_path = os.path.join(stimuli_path, "Trajectory_0{}_central".format(dir_id))
        for file_id in range(700):
            image_path = os.path.join(stim_path, "Trajectory_0{}_{}.png".format(dir_id, file_id))
            print(image_path)
            image = imageio.imread(image_path)
#             resized_image = cv2.resize(image, (64,64), interpolation=cv2.INTER_CUBIC)
            stimuli.append(image)
        all_stimuli.append(np.array(stimuli))
    return np.array(all_stimuli)
```

```python
stimuli = load_images_stimuli(CAT_STIMULI_PATH)
```
